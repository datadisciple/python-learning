```python
import pandas as pd
```

```python
# assign an na_values list to pass into read_csv() as argument so df knows what values it can treat as NaN
na_vals = ['NA', 'Missing']
df = pd.read_csv('../data/survey_results_public.csv', index_col='Respondent', na_values=na_vals)
schema_df = pd.read_csv('../data/survey_results_schema.csv', index_col='Column')
```

```python
df['YearsCode'].head(10)
```

```python
# cannot calculate mean on Series due to NaN values
df['YearsCode'].mean()
```

```python
# this also doesn't work because there is a string data type of value 'Less than 1 year'
df['YearsCode'].astype(float).mean()
```

```python
# view all unique values in column to see what strings are mixed in
df['YearsCode'].unique()
```

```python
# replace < 1 yr with 0
df['YearsCode'].replace('Less than 1 year', 0, inplace=True)
```

```python
# replace > 50 yrs with 51
df['YearsCode'].replace('More than 50 years', 51, inplace=True)
```

```python
# can know convert from int to float
df['YearsCode'] = df['YearsCode'].astype(float)
```

```python
# mean is know able to be calculated as all data types are floats
df['YearsCode'].mean()
```
