```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Adam'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Doe'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', 'A@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# sorts values by last name in descending order
df.sort_values(by='last', ascending=False)
```

```python
# can sort on multiple columns
df.sort_values(by=['last', 'first'], ascending=False)
```

```python
# using a list to specify multiple sorting options on multiple columns
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
df
```

```python
# index gets re-sorted/reset to align with where the row is in the dataframe
# can use inplace=True argument if you want this to carry through
df.sort_index()
```

```python
# sorts values of 'last' col in ascending order
df['last'].sort_values()
```
