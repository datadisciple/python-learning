```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['email']
```

```python
# sets the email as the index
# usually you want to ensure that you are setting your index to a unique value
df.set_index('email', inplace=True)
```

```python
# because inplace was specified above the index changes carry over to future calls to the df
df
```

```python
# look at the index values
df.index
```

```python
# since the df is now index by email, an email can be passed in to loc get data for that user
df.loc['CoreyMSchafer@gmail.com', 'last']
```

```python
df.iloc[0]
```

```python
# resets the index and then uses inplace=True to ensure this change carries over to future calls to the df
df.reset_index(inplace=True)
df
```
